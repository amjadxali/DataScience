````python
import matplotlib.pyplot as plt
import pandas as pd

# Example data (replace with real dataset)
data = {
    'Year': [2005, 2006, 2007, 2008, 2009, 2010],
    'Cars': [500, 550, 600, 450, 400, 500],
    'Trucks': [300, 320, 310, 250, 240, 300],
    'SUVs': [400, 420, 430, 380, 350, 400]
}
df = pd.DataFrame(data)

# Shading recession periods
recession_periods = [(2008, 2009)]

# Plotting
plt.figure(figsize=(10, 6))
for col in df.columns[1:]:
    plt.plot(df['Year'], df[col], label=col)
for start, end in recession_periods:
    plt.axvspan(start, end, color='gray', alpha=0.3, label='Recession')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales Trends of Vehicle Types During Recession Periods')
plt.legend()
plt.show()
````

````python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example dataset
# Replace this with your actual dataset
data = {
    'Date': pd.date_range(start='2020-01-01', periods=24, freq='ME'),  # Updated freq to 'ME'
    'Sales': [100, 120, 90, 110, 130, 85, 95, 105, 115, 125, 80, 100,
              150, 170, 140, 160, 180, 135, 145, 155, 165, 175, 130, 150],
    'Vehicle_Type': ['Car', 'Truck', 'Car', 'Truck', 'Car', 'Truck',
                     'Car', 'Truck', 'Car', 'Truck', 'Car', 'Truck',
                     'Car', 'Truck', 'Car', 'Truck', 'Car', 'Truck',
                     'Car', 'Truck', 'Car', 'Truck', 'Car', 'Truck'],
    'Recession': ['Yes', 'Yes', 'Yes', 'Yes', 'Yes', 'Yes',
                  'No', 'No', 'No', 'No', 'No', 'No',
                  'Yes', 'Yes', 'Yes', 'Yes', 'Yes', 'Yes',
                  'No', 'No', 'No', 'No', 'No', 'No']
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Create a line plot using Seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=df,
    x='Date',
    y='Sales',
    hue='Vehicle_Type',
    style='Recession',
    markers=True,
    dashes=False
)

# Customize the plot
plt.title('Sales Trend per Vehicle Type: Recession vs Non-Recession Periods', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.legend(title='Vehicle Type and Recession Status')
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
````

### TASK 1.4: Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.
<br>Now, you want to find more insights from the data to understand the reason. <br>Plot a two line charts using subplotting to answer:-
#### How did the GDP vary over time during recession and non-recession periods? 
<br>Make use of <u>add_subplot()</u> from Matplotlib for this comparision.


````python
import matplotlib.pyplot as plt
import pandas as pd

# Example data
data = {
    "Year": [2000, 2001, 2002, 2003, 2004, 2005],
    "GDP": [2.5, 1.2, 0.5, 3.0, 3.5, 4.0],
    "Recession": [True, True, False, False, False, True]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Separate data into recession and non-recession periods
recession = df[df["Recession"] == True]
non_recession = df[df["Recession"] == False]

# Create subplots
fig, ax = plt.subplots(2, 1, figsize=(10, 8))

# Plot GDP during recession
ax[0].plot(recession["Year"], recession["GDP"], marker="o", color="red", label="Recession")
ax[0].set_title("GDP During Recession")
ax[0].set_xlabel("Year")
ax[0].set_ylabel("GDP")
ax[0].legend()

# Plot GDP during non-recession
ax[1].plot(non_recession["Year"], non_recession["GDP"], marker="o", color="green", label="Non-Recession")
ax[1].set_title("GDP During Non-Recession")
ax[1].set_xlabel("Year")
ax[1].set_ylabel("GDP")
ax[1].legend()

# Adjust layout
plt.tight_layout()
plt.show()
````

 ### How does the average vehicle price relate to the sales volume during recessions?
 <br> Plot another scatter plot and title it as 'Relationship between Average Vehicle Price and Sales during Recessions'


````python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Check the current working directory
print("Current working directory:", os.getcwd())

# Provide the correct file path
file_path = 'DA0101EN/labs/Module 6/housing.csv'  # Specify the actual path of your CSV file
if not os.path.exists(file_path):
    print(f"File not found: {file_path}")
    print("Please ensure the file path is correct.")
else:
    # Load the data
    data = pd.read_csv(file_path)

    # Ensure your data has the necessary columns
    try:
        data['date'] = pd.to_datetime(data['date'])
    except KeyError:
        print("Column 'date' is missing in your data. Please check your file.")
    else:
        # Filter data for recession periods
        recession_start = '2008-01-01'
        recession_end = '2009-12-31'
        recession_data = data[(data['date'] >= recession_start) & (data['date'] <= recession_end)]

        # Check if necessary columns exist for correlation
        if 'vehicle_price' not in recession_data.columns or 'sales_volume' not in recession_data.columns:
            print("Required columns 'vehicle_price' or 'sales_volume' are missing. Please check your data.")
        else:
            # Calculate correlation
            correlation, _ = pearsonr(recession_data['vehicle_price'], recession_data['sales_volume'])
            print(f"Correlation between vehicle price and sales volume during the recession: {correlation}")

            # Plot scatter plot
            plt.figure(figsize=(10, 6))
            plt.scatter(recession_data['vehicle_price'], recession_data['sales_volume'], alpha=0.6, edgecolors='k')
            plt.title('Vehicle Price vs Sales Volume During Recession')
            plt.xlabel('Vehicle Price')
            plt.ylabel('Sales Volume')
            plt.grid(True)
            plt.show()
````
